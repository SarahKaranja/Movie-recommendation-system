```python
import pandas as pd
```

```python
df1 = pd.read_csv('links.csv')
df1
```

```python
df2 = pd.read_csv('movies.csv')
df2
```

```python
df3 = pd.read_csv('ratings.csv')
df3
```

```python
df4 = pd.read_csv('tags.csv')
df4
```

```python
# Merge df1 and df2 on 'movieId' column
merged_df = pd.merge(df1, df2, on='movieId', how='inner')

# Merge merged_df and df3 on 'movieId' column
merged_df = pd.merge(merged_df, df3, on='movieId', how='inner')

# Merge merged_df and df4 on 'movieId' column
merged_df = pd.merge(merged_df, df4, on='movieId', how='inner')

# Display the merged DataFrame
merged_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Define mapping of genres to broader categories
genre_mapping = {
    'Action': ['Action', 'Adventure'],
    'Comedy': ['Comedy'],
    'Drama': ['Drama'],
    'Horror': ['Horror', 'Thriller'],
    'Sci-Fi': ['Sci-Fi', 'Fantasy']
    # Add more mappings as needed
}

# Function to map genres to broader categories
def map_to_broad_category(genre):
    for category, genres in genre_mapping.items():
        if genre in genres:
            return category
    return 'Other'

# Apply mapping to create a new column with broader categories
merged_df['broad_genre'] = merged_df['genres'].apply(map_to_broad_category)

# Create count plot with binned genres
plt.figure(figsize=(10, 6))
sns.countplot(x='broad_genre', data=merged_df, palette='viridis')
plt.title('Frequency of Movies by Broad Genre')
plt.xlabel('Broad Genre')
plt.ylabel('Frequency')
plt.show()
```
